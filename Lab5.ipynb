```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

```python
Out_Time = np.loadtxt('lab_5_data.csv', delimiter=',', skiprows=1,usecols=0)
Out_Particles = np.loadtxt('lab_5_data.csv', delimiter=',', skiprows=1,usecols=1)
Out_Temp = np.loadtxt('lab_5_data.csv', delimiter=',', skiprows=1,usecols=2)
```

```python
In_Time = np.loadtxt('lab_5_data_inside.csv', delimiter=',', skiprows=1,usecols=0)
In_Particles = np.loadtxt('lab_5_data_inside.csv', delimiter=',', skiprows=1,usecols=1)
In_Temp = np.loadtxt('lab_5_data_inside.csv', delimiter=',', skiprows=1,usecols=2)
```

```python
#Particles/Time
plt.figure(figsize=(8,8))
plt.plot(Out_Time,Out_Particles, label="Particles/Time Outside",linestyle='--', color='b')
plt.xlabel(xlabel="Time (seconds)")
plt.ylabel(ylabel="Particles (2.5um/0.1L air)")

plt.plot(In_Time,In_Particles, label="Particles/Time Inside",linestyle='--', color='g')
plt.xlabel(xlabel="Time (seconds)")
plt.ylabel(ylabel="Particles (2.5um/0.1L air)")

plt.legend(loc=2,bbox_to_anchor=(-.2, 1.1))

plt.twinx()
#Temperature/Time
plt.plot(Out_Time,Out_Temp, label="Temp/Time Outside", color='b')
plt.xlabel(xlabel="Time (seconds)")
plt.ylabel(ylabel="Temerature (Celcius)")

plt.plot(In_Time,In_Temp, label="Temp/Time Inside", color = 'g')
plt.xlabel(xlabel="Time (seconds)")
plt.ylabel(ylabel="Temerature (Celcius)")

plt.legend(loc=1, bbox_to_anchor=(1.2, 1.1))
```

```python
#Particles Hist
plt.hist([In_Particles,Out_Particles],stacked=True,bins=2,color=['g','b'],label=['Outside Particles', 'Inside Particles'])
plt.xticks(ticks=(0,1))
plt.xlabel(xlabel='Number of particles with a size of 2.5 um per second')
plt.ylabel(ylabel='Count')
plt.title(label="Number of Particles Inside vs Outside")
plt.legend()
```

```python
#Temperature Hist
plt.hist([In_Temp,Out_Temp],stacked=True,bins=10,color=['g','b'],label=['Outside Temperature', 'Inside Temperature'])
plt.xlabel(xlabel='Temperature')
plt.ylabel(ylabel='Count')
plt.title(label="Number of Particles Inside vs Outside")
plt.legend()
```

```python
# Data Analysis
# 1. Mean
#Particles mean
mean_In_Particles=np.mean(In_Particles)
mean_Out_Particles=np.mean(Out_Particles)
print(f"The mean of the number of inside particles is {mean_In_Particles}")
print(f"The mean of the number of outside particles is {mean_Out_Particles}")
#Particles std
std_In_Particles=np.std(In_Particles)
std_Out_Particles=np.std(Out_Particles)
print(f"The mean standard deviation of the number of inside particles is {std_In_Particles/(30**(1/2))}")
print(f"The mean standard deviation of the number of outside particles is {std_Out_Particles/(30**(1/2))}")

#Temperature mean
mean_In_Temp=np.mean(In_Temp)
mean_Out_Temp=np.mean(Out_Temp)
print(f"The mean of the temperature inside is {mean_In_Temp}")
print(f"The mean of the temperature outside is {mean_Out_Temp}")
#Temperature std
std_In_Temp=np.std(In_Temp)
std_Out_Temp=np.std(Out_Temp)
print(f"The mean standard deviation of temperature inside {std_In_Temp/(30**(1/2))}")
print(f"The mean standard deviation of temperature outside {std_Out_Temp/(30**(1/2))}")
```
