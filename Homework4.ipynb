```python
# Data Analysis
# 1. Mean
#Particles mean
mean_In_Particles=np.mean(In_Particles)
mean_Out_Particles=np.mean(Out_Particles)
print(f"The mean of the number of inside particles is {mean_In_Particles}")
print(f"The mean of the number of outside particles is {mean_Out_Particles}")
#Particles std
std_In_Particles=np.std(In_Particles)
std_Out_Particles=np.std(Out_Particles)
print(f"The mean standard deviation of the number of inside particles is {std_In_Particles/(299**(1/2))}")
print(f"The mean standard deviation of the number of outside particles is {std_Out_Particles/(299**(1/2))}")

#Temperature mean
mean_In_Temp=np.mean(In_Temp)
mean_Out_Temp=np.mean(Out_Temp)
print(f"The mean of the temperature inside is {mean_In_Temp}")
print(f"The mean of the temperature outside is {mean_Out_Temp}")
#Temperature std
std_In_Temp=np.std(In_Temp)
std_Out_Temp=np.std(Out_Temp)
print(f"The mean standard deviation of temperature inside {std_In_Temp/(299**(1/2))}")
print(f"The mean standard deviation of temperature outside {std_Out_Temp/(299**(1/2))}")
```
